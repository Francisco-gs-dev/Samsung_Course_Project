# Extracción de datos a futuro. (En este caso sólo de una provincia, pero se podría iterar para conseguir el de varias provincias y municipios)

```python
import requests
from bs4 import BeautifulSoup
import csv

urls = []
for i in range(10):
    url = f'https://www.yr.no/en/forecast/hourly-table/2-2521978/Spain/Valencia/Alicante/Alicante?i={i}'
    urls.append(url)


resumen = []


for url in urls:

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        fecha_element = soup.find('h2', class_='header-3 heading--color-primary')
        if fecha_element:
            fecha = fecha_element.find('time')
            if fecha:
                fecha_texto = fecha.text.strip()
            else:
                continue
        else:
            print(f"No se encontró el elemento <h2> con la fecha en la URL {url}")
            continue

        filas = soup.find_all('tr')

        for fila in filas[1:]:
            celdas = fila.find_all('td')
            if len(celdas) > 3:
                hora = celdas[0].text.strip()
                temp = celdas[2].text.strip()
                viento = celdas[4].text.strip()

                resumen.append([fecha_texto, hora, temp, viento])
        else:
            print(f"No se encontró la fecha en la URL {url}")

    else:
        print(f"No se pudo obtener la URL {url}, código de estado: {response.status_code}")

encabezados = ['Fecha', 'Hora', 'Temperatura', 'Viento']
with open('resultados_yr_semanales.csv', mode='w', newline='') as archivo:
    writer = csv.writer(archivo)
    writer.writerow(encabezados)
    writer.writerows(resumen)
```
